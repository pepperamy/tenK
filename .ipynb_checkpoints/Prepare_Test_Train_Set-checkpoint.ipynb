```python
import pandas as pd
import numpy as np
import pickle
from sklearn.model_selection import train_test_split
```

```python
def load_data(data_name):
    f = open(data_name,'rb')
    data = pickle.load(f)
    return data
```

```python
data = load_data('/home/jujun/fraudprediction_10k/data/data_20200212_2012')
```

```python
text_para = []
labels = []
for d in data:
    text_para.append(d[0])
    labels.append(d[1])
```

```python
indices = np.arange(len(labels))
```

```python
text_para_1 = np.asarray(text_para)
```

```python
labels_1 = np.asarray(labels)
```

```python
X_train, X_test, y_train, y_test, indices_train, indices_test = train_test_split(text_para_1, labels_1, indices, test_size=0.2, 
                                                                                 random_state=42, stratify=labels_1)
```
